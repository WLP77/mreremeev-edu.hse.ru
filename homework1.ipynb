Задание 1

```python
year = int(input("Введите год: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Високосный год")
else:
    print("Обычный год")
```

Задание 2

```python
number = input("Введите шестизначный номер билета: ")

if len(number) == 6 and number.isdigit():
    first_half = sum(int(digit) for digit in number[:3])
    second_half = sum(int(digit) for digit in number[3:])
    
    if first_half == second_half:
        print("Счастливый билет")
    else:
        print("Несчастливый билет")
else:
    print("Ошибка: введите шестизначное число")
```
