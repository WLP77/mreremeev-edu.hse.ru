```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

def get_owner_by_doc_number(docs, doc_number):
    for doc in docs:
        if doc.get('number') == doc_number:
            return doc.get('name')
    return None

def get_shelf_by_doc_number(dirs, doc_number):
    for shelf, numbers in dirs.items():
        if doc_number in numbers:
            return shelf
    return None

def handle_command_p(docs):
    doc_number = input("Введите номер документа:\n").strip()
    owner = get_owner_by_doc_number(docs, doc_number)
    if owner:
        print(f"Владелец документа: {owner}")
    else:
        print("Владелец документа: владелец не найден")

def handle_command_s(dirs):
    doc_number = input("Введите номер документа:\n").strip()
    shelf = get_shelf_by_doc_number(dirs, doc_number)
    if shelf:
        print(f"Документ хранится на полке: {shelf}")
    else:
        print("Документ не найден")

def run_cli(docs, dirs):
    while True:
        cmd = input("Введите команду:\n").strip().lower()
        if cmd == 'q':
            print("Выход из программы.")
            break
        elif cmd == 'p':
            handle_command_p(docs)
        elif cmd == 's':
            handle_command_s(dirs)
        else:
            print("Неизвестная команда. Доступные: p, s, q.")

run_cli(documents, directories)
```
