Задание 1

```python
word = input("Введите слово: ")

n = len(word)

if n % 2 == 1:
    middle = word[n // 2]
else:
    middle = word[n // 2 - 1 : n // 2 + 1]

print("Результат:", middle)
```

Задание 2

```python
def make_pairs():
    boys = input("Введите имена юношей через запятую: ").split(",")
    girls = input("Введите имена девушек через запятую: ").split(",")

    boys = [b.strip() for b in boys if b.strip()]
    girls = [g.strip() for g in girls if g.strip()]

    if len(boys) != len(girls):
        print("Внимание, кто-то может остаться без пары.")
        return

    boys_sorted = sorted(boys)
    girls_sorted = sorted(girls)

    print("Идеальные пары:")
    for b, g in zip(boys_sorted, girls_sorted):
        print(f"{b} и {g}")

make_pairs()
```
